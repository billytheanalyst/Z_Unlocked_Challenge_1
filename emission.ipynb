```python
# Manipulation Libraries
import numpy as np
import pandas as pd

# Visualization Libraries
import plotly.graph_objects as go
```

```python
# Import Data
energy = pd.read_csv("energy_use_data_11-29-2021.csv")
energy.describe()
```

```python
energy['Item'].value_counts()
```

```python
# Import the data
areas_to_keep = ['China', 'Germany', 'Japan', 'United States of America']

evolution = energy[(energy['Area'].isin(areas_to_keep))&
                   (energy['Year']==2019)].reset_index()

evolution['Area'] = evolution['Area'].replace(to_replace=['United States of America'],
                                              value=['United States'])

evolution = evolution.groupby(['Area', 'Item'])['Value'].mean().reset_index()

totals = evolution.groupby('Area')['Value'].sum().reset_index()
evolution = pd.merge(evolution, totals, on = 'Area')
evolution['Percentage'] = evolution['Value_x'] / evolution['Value_y']
evolution['Percentage'] = evolution['Percentage'].apply(lambda x: round(x*100, 2))

evolution.head()
```
